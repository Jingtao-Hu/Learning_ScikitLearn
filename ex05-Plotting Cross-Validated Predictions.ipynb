```python
#load dataset
from sklearn import datasets
boston = datasets.load_boston()
X = boston.data
y = boston.target
```

```python
print(boston.data.shape)
print(boston.feature_names.shape)
print(boston.target.shape)
```

```python
#estimator
from sklearn import linear_model
lr = linear_model.LinearRegression()

#another estimator
from sklearn import svm
clf = svm.SVC().fit(X,y)
```

```python
#model
from sklearn.model_selection import cross_val_predict
pred = cross_val_predict(lr, X, y, cv=10)
pred.shape
```

```python
from sklearn.model_selection import train_test_split
splitedData = train_test_split(X, y,test_size=0.3)
X_train = splitedData[0]
X_test = splitedData[1]
y_train = splitedData[2]
y_test = splitedData[3]
print("The shape of X_train:",X_train.shape)
print("The shape of X_test:",X_test.shape)
print("The shape of y_train:",y_train.shape)
print("The shape of y_test:",y_test.shape)
```

```python
#image
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter(y,pred,color='red',edgecolors=(0,0,0))
ax.plot([y.min(),y.max()],[y.min(),y.max()],'k--',lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
