```python
iris = datasets.load_iris()
iris.data.shape
```

```python
print(iris.data)
print(iris.feature_names)
print(iris.target)
print(iris.target_names)
```

```python
import pandas as pd
```

```python
data = iris.data
feature_names = iris.feature_names
target = iris.target
target_names = iris.target_names

data_with_feature = {feature_names[0] : data[:,0],feature_names[1] : data[:,1],feature_names[2] : data[:,2], "target":target}
    
data_with_feature = pd.DataFrame(data_with_feature)

data_with_feature['PetalLengthCm'] = data_with_feature['petal length (cm)']
data_with_feature['SepalLengthCm'] = data_with_feature['sepal length (cm)']
data_with_feature['SepalWidthCm'] = data_with_feature['sepal width (cm)']
data_with_feature.drop(['petal length (cm)','sepal length (cm)', 'sepal width (cm)'],axis=1,inplace=True)
data_with_feature.drop('target',axis=1,inplace=True)
data_with_feature.insert(3,'Target',target)
data_with_feature
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()

from sklearn.model_selection import train_test_split
trainSet, testSet = train_test_split(data_with_feature, test_size=0.33)
print(trainSet.shape)
print(testSet.shape)
print(trainSet.head(3))
```

```python
trainData = pd.DataFrame.as_matrix(trainSet[['PetalLengthCm','SepalLengthCm','SepalWidthCm']])
trainTarget = pd.DataFrame.as_matrix(trainSet[['Target']]).ravel() #change vector to series
testData = pd.DataFrame.as_matrix(testSet[['PetalLengthCm','SepalLengthCm','SepalWidthCm']])
testTarget = pd.DataFrame.as_matrix(testSet[['Target']]).ravel()
```

```python
clf.fit(trainData, trainTarget)
y_pred = clf.predict(testData)
ground_truth = testTarget
print(y_pred)
print(ground_truth)
```

```python
import numpy as np
correct_prediction = np.equal(y_pred,ground_truth).sum()
accuracy = correct_prediction/len(testData)
print("Accuracy is:", accuracy)
```

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001,C=100.)
clf.fit(trainData, trainTarget)
y_pred = clf.predict(testData)
ground_truth = testTarget
print(y_pred)
print(ground_truth)
correct_prediction = np.equal(y_pred,ground_truth).sum()
accuracy = correct_prediction/len(testData)
print("Accuracy is:", accuracy)
```
